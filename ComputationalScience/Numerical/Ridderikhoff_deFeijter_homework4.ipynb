# Exercise 1
# (a)
Let $Q$ be a real, square orthogonal matrix. Show that all eigenvalues of $Q$ satisfy $|\lambda |=1$.

We know that $Qx = \lambda x$ by definition of the eigenvalues. From there we can state $||Qx||^{2} = ||\lambda x||^{2} = |\lambda|^{2} ||x||^{2}$. 
Since Q is a square matrix, $||Qx||^{2} = (Qx)^{T}(Qx)$, so $(Qx)^{T}(Qx) = |\lambda|^{2}(x)^{T}(x)$. Since Q is orthogonal, $Q^{T} Q = I$, so $(x)^{T}(x) = |\lambda|^{2}(x)^{T}(x)$. This means that $|\lambda|^{2}$ must be 1, so $|\lambda| = 1$. 


# (b)
A projection matrix is a matrix $P$ for which $P^2=P$. Find the eigenvalues of a projection matrix.

We know that $Px = \lambda x$. Since $P = P^{2}$, $Px = \lambda x = P^{2}x = \lambda^{2}x$. So we see that $\lambda x = \lambda^{2}x$, which means that $\lambda = \lambda^{2}$. This can only be true for 2 values: 0 and 1.

# Exercise 2
Generate a symmetric $4 × 4$ matrix $A$ randomly, e.g. as below.
Implement Rayleigh quotient iteration (Algorithm 4.4) to compute the largest eigenvalue and corresponding eigenvector of $A$. Use a suitable stopping criterion for your iterations (and explain why you think it is suitable). You may use `scipy.linalg.solve`.

Example 4.14! 

```python
import numpy as np

# create matrix
B = np.random.randn(4,4)
A = B+B.T

# use built-in function to check real eigenvalues
echteeig = np.linalg.eig(A)[0]

# first approximation of largest eigenvalue
mu = 10

# first approximation of corresponding eigenvector
x0 = np.array([1, 1, 1, 1])

# perform power iteration to improve initial eigenvector
for i in list(range(0, 5)):
    x0 = np.dot(A, x0)
    
x = [x0]
mus = [mu]

# create identity matrix
I = np.identity(4)

stop = False
    
# perform Rayleigh quotient iteration (with result of power iteration)
while(stop == False):
# for i in list(range(0, 5)):
    
    # compute shift of eigenvalue approximation
    prod1 = np.dot(x[-1].transpose(), A)
    prod2 = np.dot(prod1, x[-1])
    prod3 = np.dot(x[-1].transpose(), x[-1])
    nextMu = prod2/prod3
    
    # generate next corresponding eigenvector
    nextX = np.linalg.solve(A - nextMu * I, x[-1])
    
    # normalize eigenvector
    nextX = nextX/np.linalg.norm(nextX)
    
    x.append(nextX)
    mus.append(nextMu)
    
    # stop iterating if no (almost) improvement is made compared to last eigenvalue approximation
    if len(mus)> 1 and abs(mus[-1] - mus[-2]) < 10**-15:
        stop = True
    
print "Built-in maximum absolute eigenvalue calculation: ", max(echteeig, key=abs)
print "Our Rayleigh quotient iteration approximation of largest eigenvalue: ", mus[-1]
```

# Exercise 3
Let $A$ be a $n × n$ tridiagonal matrix that results (up to an overall constant scaling factor) from an
$n$-point spatial discretization of the Laplace operator in $1$ spatial dimension. Specifically, all diagonal
elements of $A$ have the value $-2$, all elements just below or above the diagonal have value $1$.

$${\displaystyle A = \left( \begin{array}{rrrrr} -2 &  1  &  0 &  0& \ldots \\ 1  & -2 & 1 & 0 &\ldots   \\ 0 & 1 & -2 &  1 &\ldots \\ \vdots & & & & \ddots \end{array} \right) \,}$$



Take $n = 150$. Use `scipy.linalg.qr` to implement Algorithm 4.7 from the book and compute the full eigenvalue spectrum of $A$. Compare against
the eigenvalues obtained by the built-in routine `scipy.linalg.eig(A)`. Show in a figure the (approximate)
eigenvalues after 10, 100 and 500 QR iterations, as well as the eigenvalues obtained with `eig(A)`.


```python
from scipy.sparse import diags
from scipy.linalg import qr, eig
import matplotlib.pyplot as plt

# create the tridiagonal matrix
originalA = diags([1, -2, 1], [-1, 0, 1], shape=(150, 150)).toarray()
A10 = originalA
A100 = originalA
A500 = originalA

# perform 10 QR iterations
for i in list(range(0, 10)):
    Q, R = qr(A10)
    A10 = np.dot(R, Q)
    
# perform 100 QR iterations
for i in list(range(0, 100)):
    Q, R = qr(A100)
    A100 = np.dot(R, Q)
    
# perform 500 QR iterations
for i in list(range(0, 500)):
    Q, R = qr(A500)
    A500 = np.dot(R, Q)

# take values from the diagonals of the iterated matrices
eigenA10 = []
eigenA100 = []
eigenA500 = []
for i in list(range(0, 150)):
    eigenA10.append(A10[i, i])
    eigenA100.append(A100[i, i])
    eigenA500.append(A500[i, i])    

# plot computed eigenvalues approximations
fig, ax = plt.subplots()
ax.plot(list(range(0, 150)), eigenA10, label = "10 it QR")
ax.plot(list(range(0, 150)), eigenA100, label = "100 it QR")
ax.plot(list(range(0, 150)), eigenA500, label = "500 it QR")
ax.plot(list(range(0, 150)), eig(originalA)[0], label = "Built-in routine")
ax.set_xlim(0, 150)
ax.set_xlabel("Number of eigenvalue")
ax.set_ylabel("Value of eigenvalue")
ax.legend(fontsize = "medium", loc = 2)
plt.show()
    
```

As can be seen in this figure, the number of QR iterations is of influence on the accuracy of the calculation of the eigenvalues on the diagonal. After only 10 iterations, the eigenvalues are very inaccurate, whereas after 100 iterations the basic shape of eigenvalues is more recognizable. After 500 QR iterations the eigenvalues mostly overlap with the built-in routine of scipy.
