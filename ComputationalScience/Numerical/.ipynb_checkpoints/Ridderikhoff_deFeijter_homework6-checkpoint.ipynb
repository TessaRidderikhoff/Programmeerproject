# Exercise 1
Let $P_k$ be the $k$th _Legendre polynomial_ (see paragraph 7.3.4) on $[-1, 1]$.

We know that $P_0(x) = 1$, $P_1(x) = x$, and that $P_k$ satisfy the recurrence relation 

$$k P_k(x) =x (2k-1)  P_{k-1}(x) - (k-1) P_{k-2}(x).$$

We will use [_induction_](https://en.wikipedia.org/wiki/Mathematical_induction) to show that the derivatives of the Legendre polynomials satisfy

$$P_k'(1) = k(k+1)/2.$$

## (a) Base case
Show that $P_0'(1) = 0$ and $P_1'(1) = 1$.

$P_{1}'(1) = 1(1+1)/2 = 1$ and $P_{0}'(1) = 0(0+1)/2 = 0$. 

## (b)  Induction Step
Let $k \geq 2$. Use the recurrence relation to show that $P_k'(1) = k(k+1)/2$. Assume $P_{k}(1) = 1$.

Using the product rule to find the derivitive of $P_{k}(x) = (x(2k - 1)P_{k-1}(x) - (k-1)P_{k-2}(x)) * \frac{1}{k}$, we find that $P_{k}'(x) = (((2k - 1)P_{k-1}(x)) + ((2kx - x)P_{k-1}'(x)) - (k-1)P_{k-2}'(x)) * \frac{1}{k}$. 

If we check this with $k = 2$: $P_{2}'(x) = (3 * P_{1}(x) + (4x - x)P_{1}'(x) - P_{0}'(x)) * \frac{1}{2}$. 
Then, $P_{2}'(1) = (3 * 1 + 3 * 1 - 0) * \frac{1}{2} = 2 * (2 + 1) / 2 = 3$, so that is correct.

In general: $P_{k}'(1) = ((2k - 1) + (2k - 1)((k - 1)k/2)) - ((k - 1)((k - 2)(k - 1) / 2))) * \frac{1}{k}$. This can be reduced to $P_{k}'(1) = k(k + 1)/2$.

Where r is the growth rate. The following population counts $P_1 ,\ldots, P_8$ (in billions) are observed:

```python
data = [0.19, 0.36, 0.69, 1.3, 2.5, 4.7, 8.5, 14]
# t = [1, 2, 3, 4, 5, 6, 7, 8]
t = [0, 1, 2, 3, 4, 5, 6, 7]
```

# (a)
Read chapter 6.6 on Nonlinear Least squares. Use the Gauss-Newton Method to fit the model function $f(t, x_1, x_2) = x_1\!\cdot x_2^t$ to the data. Find estimates for the initial population $P_0=x_1$ and the growth rate $r=x_2$.

```python
import numpy as np
from numpy.linalg import inv, lstsq


def derivative_x1(x2, t):
    return -x2**t

def derivative_x2(x1, x2, t):
    return -x1 * t * x2**(t-1)

def residual(t, data, x1, x2):
    return data[t] - (x1 * x2**t)

x = np.array([1, 1])

for i in list(range(0, 10)):

    J = np.ones((8, 2))

    for i in range(0, 8):
        J[i, 0] = derivative_x1(x[1], t[i])
        J[i, 1] = derivative_x2(x[0], x[1], t[i])

    r = []
    for i in t:
        r.append(residual(i, data, x[0], x[1]))
        
    x = lstsq(J, r)[0]

#     new_x = np.dot(- np.dot(inv(np.dot(J.transpose(), J)), J.transpose()), r)
#     x = new_x
#     print x
print x
```

# Exercise 3



Runge's function is given by

$$ f(t) = 1/(1+25t^2).$$

# (a)
Interpolate Runge's function on $[-1,1]$ by a polynomial $p_k$ of degree $k$ for $k = 5,10,15$. Use equally spaced interpolation points. Draw your interpolating polynomials together with $f$.

```python
import matplotlib.pyplot as plt

t = np.linspace(-1, 1)

fig, ax = plt.subplots()
ax.plot(t, 1/(1 + 25*(t**2)))


for i in ([5, 10, 15]):
    coefficients = np.polyfit(t, 1/(1 + 25*(t**2)), i)
    polynomial = np.poly1d(coefficients)
    ax.plot(t, polynomial(t), label = i)
    
ax.legend(fontsize = "medium")
plt.show()
```

As can be seen, we polynomials fit the original function quite well in the middle (better when the degree of the polynomial is higher), but around the edges the fit becomes less accurate.
